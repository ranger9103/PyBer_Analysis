```python
#import dependencies
import matplotlib.pyplot as plt
import statistics
```

```python
#set the x-axis to a list of strings for each month
x_axis=["Jan","Feb","Mar","April","May","June","July","Aug","Sep","Oct","Nov","Dec"]

#set the y-axis to a list of floats as the total fare in US dollars accumulated each month
y_axis=[10.02, 23.24,39.20,35.42,32.34,27.04,43.82,10.56,11.85,27.90,20.71,20.09]
```

```python
stdev=statistics.stdev(y_axis)
```

```python
stdev
```

```python
plt.errorbar(x_axis,y_axis,yerr=stdev)
```

```python
plt.errorbar(x_axis,y_axis,yerr=stdev,capsize=3)
```

```python
fig, ax=plt.subplots()
ax.errorbar(x_axis,y_axis,yerr=stdev,capsize=3)
plt.show()
```

```python
plt.bar(x_axis,y_axis,yerr=stdev,capsize=3)
```

```python
fig, ax=plt.subplots()
ax.bar(x_axis,y_axis,yerr=stdev,capsize=3)
plt.show()
```

```python
import numpy as np
plt.barh(x_axis,y_axis)
plt.xticks(np.arange(0,51,step=5.0))
plt.gca().invert_yaxis()
```

```python
import numpy as np
np.arange(0,51,step=5.0)
```

```python
fig, ax=plt.subplots()
ax.barh(x_axis,y_axis)
ax.set_xticks(np.arange(0,51,step=5.0))
plt.gca().invert_yaxis()
plt.show()
```

```python
from matplotlib.ticker import MultipleLocator

#increase the size of the plot figure
fig, ax=plt.subplots(figsize=(8,8))
ax.barh(x_axis,y_axis)
ax.set_xticks(np.arange(0,51,step=5.0))

#create minor ticks at an increment of 1
ax.xaxis.set_minor_locator(MultipleLocator(1))
plt.gca().invert_yaxis()
plt.show()
```

```python
%matplotlib inline
#Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

#load in csv
pyber_ride_df=pd.read_csv("Resources/PyBer_ride_data.csv")
pyber_ride_df
```

```python
pyber_ride_df.plot(x="Month",y="Avg. Fare ($USD)")
plt.show()
```

```python
# set x-axis and tick locations
x_axis=np.arange(len(pyber_ride_df))
tick_locations=[value for value in x_axis]

#plot the data
pyber_ride_df.plot(x="Month",y="Avg. Fare ($USD)")
plt.xticks(tick_locations, pyber_ride_df["Month"])
plt.show()
```
